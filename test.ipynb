```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import pandas_ta as ta
```

```python
data = pd.read_csv("data/DJIA.csv")

data["DATE"] = pd.to_datetime(data["DATE"])

data["DJIA"] = pd.to_numeric(data["DJIA"], errors="coerce")
data['EMAF']=ta.ema(data.DJIA, length=20)
data['EMAM']=ta.ema(data.DJIA, length=100)
data['EMAS']=ta.ema(data.DJIA, length=150)
```

```python
plt.figure().set_figwidth(15)
plt.figure(dpi=1200)
plt.plot(data["DJIA"])
plt.plot(data["EMAF"])
plt.plot(data["EMAM"])
plt.plot(data["EMAS"])
plt.show()
```
